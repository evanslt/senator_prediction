Simple logistic regression. Just testing out using Scikitlearn 

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline 
```

```python
from sklearn import linear_model
from sklearn.metrics import classification_report,confusion_matrix
```

```python
election=pd.read_table('Sen1.tsv')
```

```python
election
```

```python
mod=linear_model.LogisticRegression(C=1e5)
```

```python
features=election[election.columns[2:4]]
#using just two features: incumbent and held by dem

win=election["Democrat Winner"]
```

```python
#train on the 40 of the earlier elections. 

X=features[120:]
y=win.values[120:]

X_test=features[0:119]
y_test=win.values[0:119]
```

```python
mod.fit(X,y)
```

```python
mod.score(X,y)
```

```python
prediction=mod.predict(X_test)
```

```python
compare= prediction-y_test
```

```python
states=election["State"]
year=election["Year"]
```

```python
def results(slice, compare):
    incorrect=0
    for i,val in enumerate(states[0:slice]):
        if compare[i]==0:
            print(str(year[i]) + ": The prediction for "+str(val)+" was correct")
        else:
            print(str(year[i]) + ": The prediction for "+str(val)+" was WRONG")
            incorrect+=1
    print("Number incorrect: "+str(incorrect))
    
```

```python
results(119, compare)
```

```python
X2=election[election.columns[2:7]][120:]
X_test2=election[election.columns[2:7]][0:119]
```

```python
mod2=linear_model.LogisticRegression(C=1e5)
```

```python
mod2.fit(X2,y)
mod2.score(X2, y)
```

```python
prediction2=mod2.predict(X_test2)
```

```python
results(119, prediction2-y_test)
```

```python
def compare_results(slice, result1, result2):
    for i,val in enumerate(states[0:slice]):
        if result1[i]-result2[i]!=0:
            print(str(year[i]) + ": discrepency in predicting "+str(val))
        
```

```python
compare_results(119, prediction, prediction2)
```
